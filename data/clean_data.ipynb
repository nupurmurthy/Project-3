```python
import pandas as pd
import numpy as np
import sqlalchemy
from sqlalchemy import create_engine
from sqlalchemy import inspect

pd.options.display.max_rows = 999
```

```python
# The path to our CSV file
file = "population.csv"

# Read our data into pandas
df = pd.read_csv(file)
df.head()
```

```python
#change data type
#df['1960.0'] = (df['1960.0'] !='n').astype(int)
df.iloc[:,4:]=df.iloc[:,4:].astype(pd.Int64Dtype())
#df.dtypes

#drop rows
df.columns=df.iloc[2]
df.drop(df.index[[0,1,2]], inplace=True)

#drop columns
df.drop(['Indicator Name','Indicator Code'], axis=1, inplace=True)

#drop column with no value ex) 2019
df=df.loc[(df.sum(axis=1) != 0), (df.sum(axis=0) != 0)]

df.rename(columns = {'Country Name':'Country_Name','Country Code':'Country_Code' }, inplace = True) 
#df.drop(['Country_Code'], axis=1, inplace=True)

df.head(5) #263x60
#df.to_csv('population_cleaned.csv', index=False)
```

```python
file1 = "life.csv"

# Read our data into pandas
df1 = pd.read_csv(file1, skiprows=3)

#drop columns
df1.drop(['Indicator Name','Indicator Code'], axis=1, inplace=True)

#drop raw or column all=0
df1=df1.loc[(df1.sum(axis=1) != 0), (df1.sum(axis=0) != 0)]

df1.rename(columns = {'Country Name':'Country_Name','Country Code':'Country_Code'}, inplace = True)
#df1.drop(['Country_Code'], axis=1, inplace=True)

df1.head(5)  #254x59
#df1.to_csv('life_cleaned.csv', index=False)
```

```python
file2 = "gdp.csv"

# Read our data into pandas
df2 = pd.read_csv(file2, skiprows=3)

#change data type
#df2.iloc[:,4:]=df2.iloc[:,4:].astype(pd.Int64Dtype())
#df.dtypes

#drop rows
#df2.columns=df2.iloc[2]
#df2.drop(df2.index[[0,1,2]], inplace=True)

#drop columns
df2.drop(['Indicator Name','Indicator Code'], axis=1, inplace=True)

#drop raw or column all=0
df2=df2.loc[(df2.sum(axis=1) != 0), (df2.sum(axis=0) != 0)]

df2.rename(columns = {'Country Name':'Country_Name','Country Code':'Country_Code'}, inplace = True)
#df2.drop(['Country_Code'], axis=1, inplace=True)

df2.head(5)  #258x60
#df2.to_csv('gdp_cleaned.csv', index=False)
```

```python
#df_list = df.stack()
```

```python
df_sql = df.melt(id_vars=["Country_Code" , "Country_Name"], 
                 var_name="Year", 
                 value_name="Value")
df1_sql = df1.melt(id_vars=[ "Country_Code","Country_Name"], 
                 var_name="Year", 
                 value_name="Value")
df2_sql = df2.melt(id_vars=[ "Country_Code","Country_Name"], 
                 var_name="Year", 
                 value_name="Value")
```

```python
#save to csv files
#df_sql.to_csv('population_sql.csv', index=False)
#df1_sql.to_csv('life_sql.csv', index=False)
#df2_sql.to_csv('gdp_sql.csv', index=False)
```

```python
#df2.transpose().head(5)
#df_sql.dtypes  # all object
df_sql['Year'] = df_sql['Year'].astype(int)
df_sql['Value'] = pd.to_numeric(df_sql['Value'], errors='coerce')
#df_sql.dtypes

df1_sql['Year'] = df1_sql['Year'].astype(int)
df1_sql['Value'] = pd.to_numeric(df1_sql['Value'], errors='coerce')

df2_sql['Year'] = df2_sql['Year'].astype(int)
df2_sql['Value'] = pd.to_numeric(df2_sql['Value'], errors='coerce')
```

```python
#df_sql #15517
```

```python
new=df_sql.merge(df1_sql, on=('Country_Code','Country_Name','Year'), suffixes=('_Population', '_Life'), how='outer')
new.head(5) #15517
```

```python
#new1=new.merge(df2_sql, on=('Country_Code','Country_Name','Year'), suffixes=('_gdp'))
new1=new.merge(df2_sql, on=('Country_Code','Country_Name','Year'), how='outer')
new1.rename(columns={'Value':'Value_GDP'}, inplace=True)
new1.head(5) #15517
```
